```python
import numpy as np
```

```python
# исходные параметры задачи
R_k = 500  # м
r_w = 0.08  # м
h = 80  # м

mu = 1  # сП
mu *= 1e-3  

k = 1  # мД
k = k * 1e-15   # м^2

m = 0.2

c_t = 2.64 * 1e-4  # 1/атм
c_t /= 101325  # 1/Па

B = 1

p_k = 250  # атм
p_k *= 101325  # Па

q_p = 110  # м^3/сут
q_p /= 86400  # м^3/с

delta_q = 10  # м^3/сут
delta_q /= 86400  # м^3/с

delta_t = 24  # ч
delta_t *= 24 * 60 * 60  # с

t_p = 680.28  # сут
t_p *= 24 * 60 * 60  # с

N = 5  # шт
```

По условию изначально скважина долгое время ($t_p=680.28$ сут) работала с дебитом $q_p=110$ м$^3$/сут.

Установившееся забойное давление на этом режиме:

```python
p_w0 = p_k - (B * mu * q_p) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w0 /= 101325
print(f'Начальное установившееся забойное давление: {p_w0} атм')
```

```python
p_w1 = p_k - (B * mu * (q_p-delta_q)) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w1 /= 101325
print(f'Установившееся забойное давление (при дебите q1): {p_w1} атм')
```

```python
p_w2 = p_k - (B * mu * (q_p-2*delta_q)) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w2 /= 101325
print(f'Установившееся забойное давление (при дебите q2): {p_w2} атм')
```

```python
p_w3 = p_k - (B * mu * (q_p-3*delta_q)) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w3 /= 101325
print(f'Установившееся забойное давление (при дебите q3): {p_w3} атм')
```

```python
p_w4 = p_k - (B * mu * (q_p-4*delta_q)) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w4 /= 101325
print(f'Установившееся забойное давление (при дебите q4): {p_w4} атм')
```

```python
p_w5 = p_k - (B * mu * (q_p-5*delta_q)) / (2 * np.pi * k * h) * np.log(R_k / r_w)
p_w5 /= 101325
print(f'Установившееся забойное давление (при дебите q5): {p_w5} атм')
```

Далее уменьшаем дебит до 100 м$^3$/сут. Неустановившийся режим. При постоянном дебите забойное давление изменяется.

Воспользуемся формулой из "Основ и приложений теории неустановившейся фильтрации" (1952 г, Нефтяное хозяйство) В.Н. Щелкачёва (формула приближённой оценки забойного давления при постоянном дебите на неустановившемся режиме).

Формула получена из аналитического решения уравнения пьезопроводности через интегральную экспоненту.

## Попытка решить уравнение пьезопроводности с помощью преобразования Лапласа

```python
import sympy as sp
from anaflow import get_lap_inv
import matplotlib.pyplot as plt
import numpy as np
```

![Photo](well_testing_handwritten_note.jpg)

```python
s, r = sp.symbols('s, r')
```

```python
mu = 0.001
k = 1e-15
h = 80
m = 0.2

rw = 0.08
re = 500
pe = 250 * 101325

c_t = 2.64 * 1e-4 / 101325

kappa = k / (m * mu * c_t)

rates = -np.array([100, 90, 80, 70, 60]) / 86400

fig1, ax1 = plt.subplots(figsize=[10,6])
for q in rates:
    arg_r, arg_re, arg_rw = sp.sqrt(s/kappa)*r, sp.sqrt(s/kappa)*re, sp.sqrt(s/kappa)*rw
    p = (q*mu)/(2*sp.pi*k*h*rw*s*sp.sqrt(s/kappa)) * \
        (sp.besselk(0, arg_re) * sp.besseli(0, arg_r) - sp.besseli(0, arg_re) * sp.besselk(0, arg_r)) / \
            (sp.besselk(0, arg_re) * sp.besseli(1, arg_rw) - sp.besseli(0, arg_re) * sp.besselk(1, arg_rw))
    display(p)
    p = sp.lambdify([s, r], p)

    time = np.linspace(36, 90000000, 50)  # в секундах
    
    for i, ri in enumerate(np.linspace(rw, re, 1)):
        # в одном часе 3600 секунд и в одной атмосфере 101325 Па
        ax1.plot(time/3600, (get_lap_inv(p, r=ri)(time) + pe)/101325, label=f'$p\,(r={round(ri,3)} м, q={round(-q*86400,1)} м^3/сут)$')
    
    ax1.set_title('Зависимость забойного давления от времени при разных дебитах \n на неустановившемся режиме')
    ax1.set_xlabel('t, ч')
    ax1.set_ylabel('p, атм')
    ax1.legend()
    ax1.minorticks_on()
    ax1.grid(which='major', linewidth=0.3)
    ax1.grid(which='minor', linewidth=0.1)
```

```python
radius = np.linspace(rw, re, 200)

fig2, ax2 = plt.subplots(figsize=[10,6])

for i, ti in enumerate(np.linspace(36,36000000,10)):
    ax2.plot(radius, (np.squeeze([get_lap_inv(p, r=ri)(ti) for ri in radius]) + pe)/101325, label = f'$t={round(ti/3600, 1)} ч$')

ax2.set_title('Зависимость давления от радиуса в различные моменты времени (неустановившийся режим)')
ax2.set_xlabel('r, м')
ax2.set_ylabel('p, атм')
ax2.legend()
ax2.minorticks_on()
ax2.grid(which='major', linewidth=0.3)
ax2.grid(which='minor', linewidth=0.1)
```
