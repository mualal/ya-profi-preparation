```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString

# входные параметры задачи
mu = 0.7
g = 9.81
b0 = 0.01
b1 = 0.01
S = np.pi * 30**2
z_0 = 5

# равномерный список моментов времени t
t_list = np.linspace(0, 100000, 10000000)
# шаг по времени
step = (t_list[-1] - t_list[0]) / (len(t_list) - 1)

# список уровней жидкости (изначально знаем только уровень при t=0)
z = [z_0]

# реализация явного метода интегрирования (метода Эйлера)
for t in t_list[1:]:
    z.append(z[-1] - (mu * np.sqrt(2 * g) * (2 * b0 / 3 * z[-1]**(3/2) + 4 * b1 / 15 * z[-1]**(5/2))) / S * step)

t_list = [t / 86400 for t in t_list]

# построение зависимости высоты уровня жидкости от времени
plt.figure()
plt.plot(t_list, z, color='blue')
plt.plot(t_list, [z_0*0.9]*len(t_list),'--', color='red')
plt.xlabel('Время t, сут')
plt.ylabel('Высота уровня жидкости, м')
plt.title('Зависимость высоты уровня жидкости от времени')
plt.grid(which='major', linewidth=0.9, linestyle='-')
plt.minorticks_on()
plt.grid(which='minor', linewidth=0.5, linestyle='--')
```

```python
# находим время, за которое призойдёт опорожнение резервуара на 10%
l1 = LineString([(i,j) for (i,j) in zip(t_list, z)])
l2 = LineString([(i,j) for (i,j) in zip(t_list, [z_0*0.9]*len(t_list))])
intersection = l1.intersection(l2)
print('Время, сут; уровень жидкости, м; (10% опорожнение резервуара)')
intersection.coords[0]
```
